```python
df1=df[['Page total likes', 'Category', 'Post Month', 'Post Weekday']].loc[0:15]
df1
```

```python
df2=df[['Page total likes', 'Category', 'Post Month', 'Post Weekday']].loc[16:30]
df2
```

```python
df3=df[['Page total likes', 'Category', 'Post Month', 'Post Weekday']].loc[31:50]
df3
```
